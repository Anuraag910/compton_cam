```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Generate example histogram data
data = np.random.normal(loc=10, scale=2, size=1000)
hist, bins = np.histogram(data, bins=300)

# Define the desired range for the Gaussian fit
fit_range_min = 8  # Minimum value for the fit range
fit_range_max = 12  # Maximum value for the fit range

# Find the indices corresponding to the fit range
fit_range_indices = np.where((bins[:-1] >= fit_range_min) & (bins[1:] <= fit_range_max))[0]

# Extract the histogram and bin centers within the fit range
hist_fit_range = hist[fit_range_indices]
bin_centers_fit_range = (bins[:-1] + bins[1:]) / 2
bin_centers_fit_range = bin_centers_fit_range[fit_range_indices]

# Define the Gaussian function
def gaussian(x, amp, mean, stddev):
    return amp * np.exp(-(x - mean)**2 / (2 * stddev**2))

# Fit the Gaussian function to the histogram data within the fit range
p0 = [np.max(hist_fit_range), np.mean(data), np.std(data)]  # Initial guess for parameters
params, _ = curve_fit(gaussian, bin_centers_fit_range, hist_fit_range, p0=p0)

# Generate the fitted curve
fit_curve = gaussian(bin_centers_fit_range, *params)

# Plot the histogram and the fitted curve within the fit range
plt.bar(bin_centers_fit_range, hist_fit_range, width=(bins[1] - bins[0]), align='center', alpha=0.5, label='Histogram')
plt.plot(bin_centers_fit_range, fit_curve, 'r-', label='Gaussian Fit')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Generate or load your histogram datasets
datasets = [np.random.normal(loc=10, scale=2, size=1000),
            np.random.normal(loc=15, scale=3, size=2000),
            np.random.normal(loc=8, scale=1.5, size=800)]

# Loop through each dataset
for i, data in enumerate(datasets):
    # Compute the histogram
    hist, bins = np.histogram(data, bins=30)
    
    # Get the bin centers
    bin_centers = (bins[:-1] + bins[1:]) / 2
    
    # Define the Gaussian function
    def gaussian(x, amp, mean, stddev):
        return amp * np.exp(-(x - mean)**2 / (2 * stddev**2))
    
    # Fit the Gaussian function to the histogram data
    p0 = [np.max(hist), np.mean(data), np.std(data)]  # Initial guess for parameters
    params, _ = curve_fit(gaussian, bin_centers, hist, p0=p0)
    
    # Generate the fitted curve
    fit_curve = gaussian(bin_centers, *params)
    
    # Plot the histogram and the fitted curve
    plt.subplot(len(datasets), 1, i+1)
    plt.bar(bin_centers, hist, width=(bins[1] - bins[0]), align='center', alpha=0.5, label='Histogram')
    plt.plot(bin_centers, fit_curve, 'r-', label='Gaussian Fit')
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.legend()

# Adjust the spacing between subplots
plt.tight_layout()

# Show the plot
plt.show()
```

Exactacting mean and variance or deviation

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Generate or load your histogram datasets
datasets = [np.random.normal(loc=10, scale=2, size=1000),
            np.random.normal(loc=15, scale=3, size=2000),
            np.random.normal(loc=8, scale=1.5, size=800)]

# Arrays to store mean and deviation
mean_values = []
deviation_values = []

# Loop through each dataset
for i, data in enumerate(datasets):
    # Compute the histogram
    hist, bins = np.histogram(data, bins=30)
    
    # Get the bin centers
    bin_centers = (bins[:-1] + bins[1:]) / 2
    
    # Define the Gaussian function
    def gaussian(x, amp, mean, stddev):
        return amp * np.exp(-(x - mean)**2 / (2 * stddev**2))
    
    # Fit the Gaussian function to the histogram data
    p0 = [np.max(hist), np.mean(data), np.std(data)]  # Initial guess for parameters
    params, _ = curve_fit(gaussian, bin_centers, hist, p0=p0)
    
    # Extract mean and deviation
    mean_values.append(params[1])
    deviation_values.append(params[2])
    
    # Generate the fitted curve
    fit_curve = gaussian(bin_centers, *params)
    
    # Plot the histogram and the fitted curve
    plt.subplot(len(datasets), 1, i+1)
    plt.bar(bin_centers, hist, width=(bins[1] - bins[0]), align='center', alpha=0.5, label='Histogram')
    plt.plot(bin_centers, fit_curve, 'r-', label='Gaussian Fit')
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.legend()

# Adjust the spacing between subplots
plt.tight_layout()

# Show the plot
plt.show()

# Print mean and deviation values
for i in range(len(datasets)):
    print(f"Dataset {i+1}: Mean = {mean_values[i]:.2f}, Deviation = {deviation_values[i]:.2f}")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
combined_list = [(1, 3), (2, 5), (3, 7), (4, 9)]

# Separate x and y values from the combined list
x = [item[0] for item in combined_list]
y = [item[1] for item in combined_list]

# Perform linear regression
coefficients = np.polyfit(x, y, 1)
m, b = coefficients

# Generate line points
line_x = np.linspace(min(x), max(x), 100)
line_y = m * line_x + b

# Plotting the combined list
plt.plot(x, y, 'o', label='Data Points')

# Plotting the fitted line
plt.plot(line_x, line_y, label='Fitted Line')

# Adding labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Combined List with Fitted Line')

# Adding legend
plt.legend()

# Displaying the plot
plt.show()
```

```python
import numpy as np

def calculate_fwhm(x, y):
    half_max = max(y) / 2.0
    idx = np.argwhere(np.diff(np.sign(y - half_max)))[0]
    
    left_idx = idx - 1
    right_idx = idx + 1
    
    left_x = x[left_idx]
    right_x = x[right_idx]
    
    fwhm = right_x - left_x
    return fwhm

# Example data
x = np.linspace(0, 10, 100)
y = np.exp(-(x - 5)**2 / 2)
plt.plot(x,y)
fwhm = calculate_fwhm(x, y)
print("FWHM:", fwhm)
```
